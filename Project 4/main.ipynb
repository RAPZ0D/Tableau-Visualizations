## Covid Activity Dataset

```python
import pandas as pd

file_path = 'covid_activity.csv'
data = pd.read_csv(file_path)

print("Data types before conversion:")
print(data.dtypes)
for col in data.columns:
    col_type = data[col].dtype
    
    if col_type == 'float64':
        data[col] = pd.to_numeric(data[col], downcast='float')
    elif col_type == 'int64':
        data[col] = pd.to_numeric(data[col], downcast='integer')

print("\nData types after conversion:")
print(data.dtypes)
```

```python
columns_of_interest = ['COUNTY_NAME', 'PROVINCE_STATE_NAME', 'CONTINENT_NAME','COUNTY_FIPS_NUMBER','COUNTRY_ALPHA_3_CODE','COUNTRY_ALPHA_2_CODE']

cleaned_data = data.dropna(subset=columns_of_interest)
```
